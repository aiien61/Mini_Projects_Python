```python
import pandas as pd
```

```python
data: pd.DataFrame = pd.read_csv('2018_Central_Park_Squirrel_Census_-_Squirrel_Data_20240708.csv')
data
```

```python
grey_squirrels: pd.Series = data["Primary Fur Color"] == 'Gray'
grey_squirrels
```

```python
grey_squirrels_count: int = len(data[data["Primary Fur Color"] == 'Gray'])
red_squirrels_count: int = len(data[data["Primary Fur Color"] == 'Cinnamon'])
black_squirrels_count: int = len(data[data["Primary Fur Color"] == 'Black'])
```

```python
data_dict = {
    "Fur Color": ['Gray', 'Cinnamon', 'Black'],
    "Count": [grey_squirrels_count, red_squirrels_count, black_squirrels_count]
}

df = pd.DataFrame(data_dict)
df.to_csv('squirrel_count.csv', index=False)
```
